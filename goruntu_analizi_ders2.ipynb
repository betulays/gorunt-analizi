# Kenarları algılama

Kenar algılama, bilgisayarlı görmede en popüler tekniklerden biridir. 
Birçok uygulamada ön işleme adımı olarak kullanılır. 
Kenar algılama ile, ışık yoğunluğunun aniden değiştiği dijital görüntüdeki noktaları işaretleyebilirsiniz. 
Bir görüntünün özelliklerinde meydana gelen ani değişiklikler, görüntülerin temsil edildiği fiziksel dünyadaki önemli olayları veya değişiklikleri vurgulamak ister.

```python
%matplotlib inline
import sys 
import cv2 
from matplotlib import pyplot as plt
```

```python
# Convert it to grayscale:
input_file = "phuketkk.jpg"
img = cv2.imread(input_file, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
output_file = input_file[:-4] + '_gray.jpg' 
cv2.imwrite(output_file, img) 
```

```python
#Extract the height and width of the image:
h, w = img.shape 
```

```python
#Sobel filtresi, yatay ve dikey kenarları ayrı ayrı algılamak için 3 x 3 çekirdek kullanan bir kenar dedektörü türüdür:
sobel_horizontal = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5) 
sobel_vertical = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5) 
```

```python
plt.imshow(sobel_horizontal)
plt.show()
```

```python
plt.imshow(sobel_vertical)
plt.show()
```

```python
#Laplacian kenar dedektörü, her iki yönde de kenarları algılar
laplacian = cv2.Laplacian(img, cv2.CV_64F) 
```

```python
plt.imshow(laplacian)
plt.show()
```

```python
canny = cv2.Canny(img, 50, 240) 
```

```python
plt.imshow(canny)
plt.show()
```
